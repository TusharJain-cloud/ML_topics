```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df['z_score'] = (df.Cholesterol - df.Cholesterol.mean())/df.Cholesterol.std()
df.head()
```

```python
df_new = df[(df.z_score < 3) & (df.z_score > -3)]
df_new.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df_new.Sex = df_new.Sex.apply(lambda x : 1 if(x == 'M') else 0)
df_new.head()
```

```python
dummies1 = pd.get_dummies(df_new.ChestPainType)
dummies1
```

```python
dummies2 = pd.get_dummies(df_new.RestingECG)
dummies2
```

```python
dummies3 = pd.get_dummies(df_new.ExerciseAngina)
dummies3
```

```python
dummies4 = pd.get_dummies(df_new.ST_Slope)
dummies4
```

```python
merge = pd.concat([df_new,dummies1,dummies2,dummies3,dummies4], axis='columns')
merge.head()
```

```python
final = merge.drop(['ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],axis='columns')
final.head()
```

```python
final = final.drop(['TA','ST','Y','Up','z_score'],axis='columns')
final.head()
```

```python
X = final.drop('HeartDisease',axis='columns')
y = final.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_scaled_train,X_scaled_test,y_train1,y_test1 = train_test_split(X_scaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores1 = cross_val_score(RandomForestClassifier(n_estimators=5),X_scaled, y, cv=10)
np.average(scores1)
```

```python
scores_logreg = cross_val_score(LogisticRegression(),X_scaled,y,cv=10)
np.average(scores_logreg)
```

```python
scores_svc = cross_val_score(SVC(),X_scaled,y,cv=10)
np.average(scores_svc)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
scores_rf = cross_val_score(RandomForestClassifier(n_estimators=5),X_pca, y, cv=10)
np.average(scores_rf)
```

```python
scores_logreg2 = cross_val_score(LogisticRegression(),X_pca,y,cv=10)
np.average(scores_logreg2)
```

```python
scores_svc2 = cross_val_score(SVC(),X_pca,y,cv=10)
np.average(scores_svc2)
```
